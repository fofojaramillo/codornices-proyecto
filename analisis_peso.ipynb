```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
plt.xkcd()
import seaborn as sns
```

```python
df = pd.read_csv("DFpesototal.csv")
df.drop(columns=[df.columns[0]], inplace=True)
```

```python
df.head()
```

```python
df["Hora"].unique()
```

```python
df["Estado"].unique()
```

```python
sns.histplot(df["PesoTotal"])
plt.show()
```

```python
df_az = df[df["Estado"] == "AZ"]
df_nm = df[df["Estado"] == "NM"]
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
sns.kdeplot(df_az["PesoTotal"], label="Arizona")
sns.kdeplot(df_nm["PesoTotal"], label="Nuevo México")
plt.title("KDE de los pesos")
plt.legend()

plt.subplot(1,2,2)
sns.kdeplot(np.log(df_az["PesoTotal"]), label="Arizona")
sns.kdeplot(np.log(df_nm["PesoTotal"]), label="Nuevo México")
plt.title("KDE de los pesos (log)")
plt.legend()
# plt.show()
```
